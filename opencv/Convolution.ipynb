### Convolution is mainly used for feature extraction

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
    
img = cv2.imread('morp.png')
```

#### Blur

```python
kernel = np.ones((5,5),np.float32)/25

###use Kernel or Blur


dst = cv2.blur(img,(5,5)) 

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()

```

#### Gaussian Blur

```python
dst = cv2.GaussianBlur(img,(5,5),0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Median Blur

```python
dst = cv2.medianBlur(img,5)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()
```

#### Blurring for edge detection

```python
# construct the Sobel x-axis kernel:used to check vertical changes in the gradient
sobelX = np.array((
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]), dtype="int")
 


dst = cv2.filter2D(img,-1,sobelX)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# construct the Sobel y-axis kernel:used to check horizontal changes in the gradient
sobelY = np.array((
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]), dtype="int")
dst = cv2.filter2D(img,-1,sobelY)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()

```

```python
laplacian = np.array((
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]), dtype="int")
dst = cv2.filter2D(img,-1,laplacian)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('processed')
plt.xticks([]), plt.yticks([])
plt.show()
```
